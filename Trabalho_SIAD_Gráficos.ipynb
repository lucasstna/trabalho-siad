# **Manipulação dos Dados**

**Retirando os dados de outras regiões da base de dados de Acidentes**

```python
selected_states = ['DF', 'GO', 'MS', 'MT']
df_acidente = df_acidente[df_acidente['uf_acidente'].isin(selected_states)]
df_acidente
```

**Combinando as Bases de Dados de Acidentes e Tipos de Veículos**

```python
df_acidente2 = df_acidente
df_acidente_veiculo = pd.merge(df_acidente2, df_veiculo, on="num_acidente", how="left")
df_acidente_veiculo
```

**Combinando com a Base de Dados de Vítimas**

```python
df_vitima2 = df_vitima
df_acidente_veiculo_vitima = pd.merge(df_acidente_veiculo, df_vitima2, on="num_acidente", how="left")
df_acidente_veiculo_vitima
```

```python
df_acidente_veiculo_vitima = df_acidente_veiculo_vitima.drop(['num_acidente', 'chv_localidade_x', 'uf_acidente_x',
                          'ano_acidente_x', 'mes_acidente_x', 'mes_ano_acidente_x',
                          'codigo_ibge', 'dia_semana', 'fase_dia', 'cond_meteorologica',
                          'end_acidente', 'num_end_acidente', 'cep_acidente', 'bairro_acidente',
                          'km_via_acidente', 'latitude_acidente', 'longitude_acidente',
                          'hora_acidente', 'tp_rodovia', 'cond_pista', 'tp_cruzamento',
                          'tp_pavimento', 'tp_curva', 'lim_velocidade', 'tp_pista', 'ind_guardrail',
                          'ind_cantcentral', 'ind_acostamento', 'qtde_acidente', 'qtde_acid_com_obitos',
                          'qtde_envolvidos_x', 'qtde_feridosilesos_x', 'qtde_obitos_x', 'ind_veic_estrangeiro',
                          'qtde_veiculos', 'chv_localidade_y', 'data_acidente_y', 'uf_acidente_y', 'ano_acidente_y',
                          'mes_acidente_y', 'mes_ano_acidente_y', 'tp_envolvido', 'gravidade_lesao', 'equip_seguranca',
                          'ind_motorista', 'qtde_envolvidos_y', 'qtde_feridosilesos_y', 'qtde_obitos_y'], axis = 1)
df_acidente_veiculo_vitima
```

**Seleção dos 10 Tipos de Acidentes Mais Comuns**

```python
contagem_valores = df_acidente_veiculo_vitima['tp_acidente'].value_counts()
top_10_valores = contagem_valores.head(10)
print(top_10_valores.index)
```

```python
df_acidente_veiculo_vitima = df_acidente_veiculo_vitima[df_acidente_veiculo_vitima['tp_acidente'].isin(top_10_valores.index)]
df_acidente_veiculo_vitima
```

**Criação do Gráfico: Correlacão entre sexo, tipo de veículo, e faixa etária nos dez tipos de acidentes
mais comuns**

```python
atributos = ['tp_acidente', 'tipo_veiculo', 'faixa_idade', 'genero']
```

```python
ax = df_acidente_veiculo_vitima[atributos].apply(pd.Series.value_counts).T.plot(kind='bar', stacked=True, figsize=(10, 6))
ax.legend(title='Categorias', bbox_to_anchor=(1.05, 0.5), loc='center left')
plt.title('Correlacão entre Sexo, Tipo de Veículo, e Faixa Etária nos Dez Tipos de Acidentes Mais Comuns')
plt.xlabel('Atributo Verificado')
plt.ylabel('Quantidade')
plt.show()
```
